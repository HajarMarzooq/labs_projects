##### 1.❓Are there better or worse days/times in terms of sales? Plot a chart to show the result

```python
# Extracting dates from transaction date

coffee_df['hour'] = pd.to_datetime(coffee_df['transaction_time']).dt.hour
coffee_df
```

```python
coffee_df["hour"].unique()
```

```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time

def time_category(hour):
    if 6 <= hour <= 12:
        return "Morning"

    elif 13 <= hour <= 17:
        return "Noon"

    elif 18 <= hour <= 21:
        return "Evening"

    else:
        return "Night"


# use apply to apply the time category into a new column called time_category

coffee_df["time_category"] = coffee_df["hour"].apply(time_category)
coffee_df.head()
```

```python
# Are there better or worse days/times in terms of sales?

sales = coffee_df.groupby(["hour", "transaction_day"])["line_item_amount"].sum().sort_values(ascending = False)
sales.head(5).plot(kind = "bar");

# Write your findings
```

##### 3.❓Do customers who have been customers longer tend to spend more? Plot a chart to show the result (Hint : You may need to use date.today() to get todays date)

```python
# Do customers who have been customers longer tend to spend more?
```

```python
from datetime import date

today = pd.to_datetime(date.today())
coffee_df["customer_tenure"] = (today - coffee_df["customer_since"]).dt.days
```

```python
coffee_df.groupby("customer_id")[["customer_tenure", "line_item_amount"]].sum()
```
